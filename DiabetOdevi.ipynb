```python
columns_to_check = ["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]
```

```python
for col in columns_to_check:
    zero_count = (df[col] == 0).sum()
    zero_percentage = 100 * zero_count / len(df)
    print(f"{col}:{zero_count} %{zero_percentage:.2f}")
```

```python
X = df.drop("Outcome",axis=1)
y=df["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=15)
```

```python
columns_to_fill = ["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]
```

```python
medians={}
for col in columns_to_fill:
    median_value = X_train[X_train[col] !=0][col].median()
    medians[col] = median_value
    X_train[col] = X_train[col].replace(0,median_value)

for col in columns_to_fill:
    X_test[col] = X_test[col].replace(0,medians[col]) 
```

```python
X_test.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report, accuracy_score,confusion_matrix
```

```python
ada = AdaBoostClassifier()
ada.fit(X_train,y_train)
y_pred = ada.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
params = {}
classification_models = {"SVC":SVC(),"LogisticRegression":LogisticRegression(),"GaussianNB":GaussianNB(),"DecisionTreeClassifier":DecisionTreeClassifier(),"KNeighborsClassifier":KNeighborsClassifier(),"RandomForestClassifier":RandomForestClassifier()}
```

```python
for i in range(len(list(classification_models))):
    model = list(classification_models.values())[i]
    model.fit(X_train, y_train)
    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)
    print(model)
    print("-----------------------------")
    print("TEST:\n")
    print("Accuracy Score: ", accuracy_score(y_test, y_test_pred))
    print("Confusion Matrix:\n ", confusion_matrix(y_test, y_test_pred))
    print("Classification Report: ", classification_report(y_test, y_test_pred))
    print("\n")
    print("TRAIN:\n")
    print("Accuracy Score: ", accuracy_score(y_train, y_train_pred))
    print("Confusion Matrix:\n ", confusion_matrix(y_train, y_train_pred))
    print("Classification Report: ", classification_report(y_train, y_train_pred))
    print("------------------------------")
    print("\n")
```

```python
params["SVC"] = {
    "C": [0.001, 0.01, 0.1, 1, 10],
    "kernel": ["linear", "poly", "rbf", "sigmoid"],
    "degree": [2, 3, 4],       
    "gamma": ["scale", "auto"],
    "max_iter": [1000, 5000]
}
params["LogisticRegression"] = [
    {   
        "penalty": ["l2"],
        "C": [0.001, 0.01, 0.1, 1, 10],
        "solver": ["lbfgs", "saga"],
        "max_iter": [1000]
    },
    {   
        "penalty": ["l1"],
        "C": [0.001, 0.01, 0.1, 1, 10],
        "solver": ["liblinear", "saga"],
        "max_iter": [1000]
    }
]
params["DecisionTreeClassifier"] = {
    "criterion": ["gini", "entropy", "log_loss"],
    "splitter": ["best", "random"],
    "max_depth": [None, 3, 5, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": [None, "sqrt", "log2"]
}
params["KNeighborsClassifier"] = {
    "n_neighbors": [3, 5, 7, 9, 11],
    "weights": ["uniform", "distance"],
    "algorithm": ["auto", "ball_tree", "kd_tree", "brute"],
    "p": [1, 2]   
}
params["RandomForestClassifier"] = {
    "n_estimators": [100, 200, 300],
    "criterion": ["gini", "entropy", "log_loss"],
    "max_depth": [None, 5, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2"]
}
params["GaussianNB"] = {
    "var_smoothing": [1e-9, 1e-8, 1e-7, 1e-6]
}
```

```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
```

```python
for i in range(len(list(classification_models))):
    model = list(classification_models.values())[i]
    model_name = list(classification_models.keys())[i]
    grid = GridSearchCV(estimator=model,param_grid=params[model_name],scoring="accuracy",n_jobs=-1)
    grid.fit(X_train, y_train)
    y_test_pred = grid.predict(X_test)
    y_train_pred = grid.predict(X_train)
    print(model)
    print("-----------------------------")
    print("TEST:\n")
    print("Accuracy Score: ", accuracy_score(y_test, y_test_pred))
    print("Confusion Matrix:\n ", confusion_matrix(y_test, y_test_pred))
    print("Classification Report: ", classification_report(y_test, y_test_pred))
    print("\n")
    print("TRAIN:\n")
    print("Accuracy Score: ", accuracy_score(y_train, y_train_pred))
    print("Confusion Matrix:\n ", confusion_matrix(y_train, y_train_pred))
    print("Classification Report: ", classification_report(y_train, y_train_pred))
    print("------------------------------")
    print("\n")
```
